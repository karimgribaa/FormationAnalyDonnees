Voici quelques exemples de mesures statistiques et de techniques que vous pouvez appliquer :

1. Statistiques descriptives basiques :

    Compter le nombre d'observations dans chaque variable.
    
    Calculer la moyenne, la médiane et le mode de chaque variable.
    
    Calculer l'écart type et la variance pour mesurer la dispersion des données.

3. Relations entre variables :

    Calculer les corrélations entre paires de variables pour évaluer les associations linéaires.
    
    Tracer des matrices de corrélation pour visualiser les relations entre les variables.
    
    Utiliser des graphiques de dispersion (scatter plots) pour représenter les relations entre deux variables continues.

4. Tests statistiques :

    Effectuer des tests d'hypothèse tels que le test t-student ou l'analyse de variance (ANOVA) pour comparer les moyennes entre groupes.

**Représenter dans le graphique de dispersion qui représente  la relation entre les variables 'engine-size' et 'price' la droite de régression liant ces deux variables.**

```python
import numpy as np

# Extraire les données des variables 'engine-size' et 'price'
engine_size = data['engine-size']
price = data['price']

# Calculer les coefficients de la régression linéaire
coefficients = np.polyfit(engine_size, price, 1)
m = coefficients[0]  # Coefficient de pente
b = coefficients[1]  # Coefficient d'interception

# Tracer le graphique de dispersion
plt.scatter(engine_size, price)

# Tracer la droite de régression
plt.plot(engine_size, m * engine_size + b, color='red')

# Ajouter des étiquettes et un titre
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.title('Relation entre Engine Size et Price avec la droite de régression')

# Afficher le graphique
plt.show()
```

**Que représente un boxplot?**

Un graphique de boîtes (box plot) est un type de visualisation qui représente graphiquement les mesures de position et de dispersion d'une variable numérique ou continue. Il fournit une représentation compacte des principales caractéristiques d'une distribution, notamment la médiane, les quartiles, les valeurs aberrantes potentielles et la plage interquartile.

Voici les éléments clés d'un graphique de boîtes :

La médiane (ligne médiane) : C'est la valeur centrale de la distribution. Elle divise les données en deux parties égales, où 50% des observations se trouvent en dessous et 50% au-dessus de la médiane.

Les quartiles (Q1 et Q3) : Ce sont les valeurs qui divisent les données en quarts égaux. Le premier quartile (Q1) représente la valeur en dessous de laquelle se situe le premier quart des données, tandis que le troisième quartile (Q3) représente la valeur en dessous de laquelle se situe le troisième quart des données.

La plage interquartile (IQR) : C'est la plage de valeurs qui couvre le milieu des données, de Q1 à Q3. Elle est utilisée pour évaluer la dispersion des données et détecter les valeurs aberrantes potentielles.

Les moustaches (whiskers) : Elles indiquent l'étendue des données, excluant les valeurs aberrantes potentielles. Les moustaches peuvent être définies de différentes manières, mais généralement, elles s'étendent jusqu'à 1,5 fois l'IQR à partir des quartiles. Les points situés au-delà des moustaches sont considérés comme des valeurs aberrantes potentielles.

Les valeurs aberrantes (outliers) : Ce sont des observations qui se trouvent en dehors des moustaches. Elles sont souvent représentées par des points individuels sur le graphique de boîtes.

Un graphique de boîtes est utile pour visualiser la distribution et la dispersion des données, ainsi que pour identifier les valeurs aberrantes potentielles. Il permet de comparer visuellement plusieurs groupes ou catégories et d'observer leurs différences dans les mesures de position et de dispersion.

##  Correlation de Pearson et P-value

**Effectuer le test de corrélation de Pearson pour évaluer la relation entre les deux variables. 'wheel_base' et 'price'.**

```python
from scipy.stats import pearsonr

# Extraire les variables 'wheel-base' et 'price'
wheel_base = data['wheel-base']
price = data['price']

# Effectuer le test de corrélation de Pearson
correlation, p_value = pearsonr(wheel_base, price)

# Afficher les résultats
print('Coefficient de corrélation de Pearson:', correlation)
print('p-valeur:', p_value)
```

**Effectuer le test de corrélation de Pearson pour évaluer la relation entre les deux variables. 'horsepower' et 'price'.**

```python
from scipy.stats import pearsonr

# Extraire les variables 'wheel-base' et 'price'
wheel_base = data['horsepower']
price = data['price']

# Effectuer le test de corrélation de Pearson
correlation, p_value = pearsonr(wheel_base, price)

# Afficher les résultats
print('Coefficient de corrélation de Pearson:', correlation)
print('p-valeur:', p_value)
```

**Effectuer l'analyse de variance (ANOVA) pour la variable 'drive-wheels'.**

```python
from scipy.stats import f_oneway

# Effectuer l'ANOVA
grouped_data = data.groupby('drive-wheels')['price'].apply(list)
anova_result = f_oneway(*grouped_data)

# Afficher les résultats
print('Statistique de test F:', anova_result.statistic)
print('p-valeur:', anova_result.pvalue)
```

**Effectuer l'analyse de variance (ANOVA) pour les catégories 'rwd' et 'fwd' la variable 'drive-wheels'.**

```python
from scipy.stats import f_oneway

# Filtrer les données pour les catégories 'rwd' et 'fwd'
rwd_data = data[data['drive-wheels'] == 'rwd']
fwd_data = data[data['drive-wheels'] == 'fwd']

# Effectuer l'ANOVA
anova_result = f_oneway(rwd_data['price'], fwd_data['price'])

# Afficher les résultats
print('Statistique de test F:', anova_result.statistic)
print('p-valeur:', anova_result.pvalue)
```
